```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

postal_codes = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(postal_codes.content,'html.parser')
postal_codes_table = soup.find_all('table')[0]
```


### Imported pandas and beautiful soup to parse and get the table data, now creating data frame for the table

```python
import numpy as np
postal_codes = []
for row in postal_codes_table.find_all('tr'):
    heading = row.find_all('th')
    if len(heading) > 0:
        postal_codes.append((heading[0].text.strip(), heading[1].text.strip(), heading[2].text.strip()))
    cols = row.find_all('td')      
    if len(cols) > 0 :
        if cols[1].text.strip() != 'Not assigned':
            if cols[2].text.strip() == 'Not assigned':
                cols[2] = cols[1]
            postal_codes.append((cols[0].text.strip(), cols[1].text.strip(), cols[2].text.strip()))
postal_codes_array = np.asarray(postal_codes)
df = pd.DataFrame(postal_codes_array)
header = df.iloc[0]
df = df[1:]
df.columns = header
df.head(11)
```
